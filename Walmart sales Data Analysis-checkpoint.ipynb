```python
np.random.seed(42)
df['Customer_ID'] = np.random.randint(1000, 1050, size=len(df))
```

```python
snapshot_date = df['Date'].max()

rfm = df.groupby('Customer_ID').agg({
    'Date': lambda x: (snapshot_date - x.max()).days,  
    'Invoice ID': 'count',                             
    'Total': 'sum'                                     
}).reset_index()

rfm.columns = ['Customer_ID', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 4, labels=[1, 2, 3, 4])
```

```python
plt.rcParams['font.family'] = 'Segoe UI Emoji'
```

```python
def rfm_level(score):
    if score >= 9:
        return ' Top Customer'
    elif score >= 6:
        return ' Loyal'
    else:
        return ' At Risk'
```

```python
rfm['RFM_Level'] = rfm['RFM_Score'].apply(rfm_level)
```

```python
rfm.head(10)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=rfm, x='RFM_Level', hue='RFM_Level', palette='Set2', legend=False)
plt.title('Customer Segmentation by RFM Score')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```
